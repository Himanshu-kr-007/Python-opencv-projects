```python
import cv2 as cv, matplotlib.pyplot as plot, numpy as np
```

# Reading Ironman Image and Ploting Graph

```python
Ironman = cv.imread("Ironman.jpg")
plot.imshow(Ironman)
```

# Cropping Tony Stark


```python
TStark = Ironman[20:280,300:480]
plot.title("Ironman")
plot.imshow(TStark)
```

```python
TStark.shape
```

# Reading Thor Image and Ploting Graph

```python
Thor = cv.imread("Thor.jpg")
plot.imshow(Thor)
```

# Cropping Face of THOR

```python

Asgardian = Thor[0:180,300:480]
plot.title("Asgardian")
plot.imshow(Asgardian)
```

```python
Asgardian.shape
```

# Swapping Face of Ironman With Thor

```python

Ironman = cv.imread("Ironman.jpg")
Ironman[40:220,320:500] = Asgardian
cv.imshow("Image", Ironman)
cv.waitKey()
cv.destroyAllWindows()
```
